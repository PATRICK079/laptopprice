# INSIGHTS ON EDA



## Categorical Features:
During the analysis, I identified that features such as "laptop," "CPU," "GPU," "model," and "brand" contained numerous categories, making them complex to handle directly. To simplify the model, I decided to drop these features for now. However, it is important to note that depending on the business context, feature engineering on these attributes could provide valuable insights, and it might be worth revisiting them if the business requirements demand it.

## Correlation with Target Variable:
A positive observation from the analysis was that all numerical features showed a strong correlation with the target variable, which is promising. This suggests that the features are informative and not redundant, which is ideal for predictive modeling. The inclusion of these features is beneficial for the model, as they contribute significantly to understanding the target.

## Outliers in Price:
There are some extreme outliers present in the price feature, which could potentially distort the model if not handled appropriately. However, I chose to leave these outliers in the dataset, as they represent real world scenarios where laptops with extraordinary prices do exist. By keeping these outliers, the model will be better equipped to capture the full range of pricing, including extreme values, which might be important in some cases.

## Redundant Status Feature:
The "status" feature, which is categorical, shows low correlation with the target variable. In most cases, such features would be dropped to avoid adding noise to the model. However, I decided to retain it in the dataset for simplicity and to avoid losing any potentially useful context, even if its impact on the target is minimal.

## Trends in RAM and Storage vs. Price:
The relationships between RAM, storage, and price generally align with expectations. laptops with higher RAM and storage tend to have higher prices. However, there are some instances where the price is unexpectedly low despite higher RAM and storage. This could indicate the presence of an underlying factor (such as brand, model, or other features) that influences the price in ways not immediately apparent from RAM and storage alone. These anomalies would warrant further investigation in future steps to identify the potential causes.
